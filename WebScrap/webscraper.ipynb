```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
title_list = []
rating_list = []
price_list = []
```

```python
base_url = "https://books.toscrape.com/catalogue/page-{}.html"
```

```python
# Use Session → reuses connection (faster)
session = requests.Session()
```

```python
for pg_num in range(1, 51):
    response = session.get(base_url.format(pg_num))
    
    # Faster parser
    soup = BeautifulSoup(response.text, "html.parser")

    # Select each book only once
    books = soup.select("article.product_pod")

    for book in books:
        # Title
        title_list.append(book.h3.a["title"])

        # Rating
        rating_list.append(book.p["class"][1])

        # Price
        price = book.select_one("p.price_color").text[1:]   # removes £ directly
        price_list.append(price)
```

```python
# Direct DataFrame creation (faster than zip → dict → DataFrame)
df = pd.DataFrame({
    "Title": title_list,
    "Rating": rating_list,
    "Price": price_list
})
```
