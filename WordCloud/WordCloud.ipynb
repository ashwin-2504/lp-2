```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Om/Downloads/archive/IMDB Dataset.csv')
print(df.head())
```

```python
print(df.columns)
print(df['review'][0])
```

```python
import re
from wordcloud import STOPWORDS
text = ''.join(df['review'].astype(str).tolist())
textdata = re.sub(r'[^A-Za-z\s]','',text)
textdata = textdata.lower()
stopwords = set(STOPWORDS)
textdata =''.join(word for word in text.split() if word not in stopwords)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 2. Call the .generate() method with your text variable inside the parentheses
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(textdata)

# 3. Plot the generated word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("IMDB Movie Reviews Word Cloud")
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=50).generate(textdata)

# 3. Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Customized IMDB Movie Reviews Word Cloud")
plt.show()
```
