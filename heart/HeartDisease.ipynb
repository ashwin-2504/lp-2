# Heart Disease Dataset Operations
Performing the same operations on the Heart Disease dataset.

```python
# Load Heart Disease Dataset
df_heart = pd.read_csv('heart_disease.csv')
display(df_heart.head())
```

### a. Data Cleaning (Heart Disease)
Checking for and handling missing values.

```python
if 'df_heart' in locals():
    # Fill NaNs with median
    df_heart = df_heart.fillna(df_heart.median(numeric_only=True))
    # Drop duplicates
    df_heart = df_heart.drop_duplicates()
    display(df_heart.info())
```

### d. Error Correcting (Heart Disease)
Removing outliers using Z-score.

```python
if 'df_heart' in locals() and 'trestbps' in df_heart.columns:
    # Removing outliers in resting blood pressure
    mean_bps = df_heart['trestbps'].mean()
    std_bps = df_heart['trestbps'].std()
    threshold = 3
    outlier_mask = ((df_heart['trestbps'] - mean_bps).abs() > threshold * std_bps)
    df_heart = df_heart[~outlier_mask]
    print('Removed outliers. Remaining shape:', df_heart.shape)
```

### b. Data Integration (Heart Disease)
Creating a new categorical feature from a numerical one to integrate into our analysis.

```python
if 'df_heart' in locals() and 'age' in df_heart.columns:
    # Binning ages into categories
    df_heart['AgeGroup'] = pd.cut(df_heart['age'], bins=[0, 40, 60, 100], labels=['Young', 'Middle', 'Senior'])
    display(df_heart[['age', 'AgeGroup']].head())
```

### c. Data Transformation (Heart Disease)
Standardizing numerical columns.

```python
if 'df_heart' in locals() and 'chol' in df_heart.columns:
    # Min-Max Scaling on cholesterol
    min_chol = df_heart['chol'].min()
    max_chol = df_heart['chol'].max()
    df_heart['chol_scaled'] = (df_heart['chol'] - min_chol) / (max_chol - min_chol)
    display(df_heart[['chol', 'chol_scaled']].head())
```
